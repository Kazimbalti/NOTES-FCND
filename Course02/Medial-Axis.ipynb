## Medial Axis


```python
import numpy as np
import matplotlib.pyplot as plt
from grid import create_grid
from skimage.morphology import medial_axis
from skimage.util import invert
from planning2 import a_star
%matplotlib inline 
```

```python
plt.rcParams['figure.figsize'] = 12, 12
```

```python
# This is the same obstacle data from the previous lesson.
filename = 'colliders.csv'
data = np.loadtxt(filename, delimiter=',', dtype='Float64', skiprows=2)
print(data)
```

Starting and goal positions in *(north, east)*.

```python
start_ne = (25,  100)
goal_ne = (650, 500)
```

```python
# Static drone altitude (meters)
drone_altitude = 5
safety_distance = 2
```

```python
grid = create_grid(data, drone_altitude, safety_distance)
skeleton = medial_axis(invert(grid))
```

Plot the edges on top of the grid along with start and goal locations.

```python
# equivalent to
# plt.imshow(np.flip(grid, 0))

plt.imshow(grid, cmap='Greys', origin='lower')
plt.imshow(skeleton, cmap='Greys', origin='lower', alpha=0.7)
    
plt.plot(start_ne[1], start_ne[0], 'rx')
plt.plot(goal_ne[1], goal_ne[0], 'rx')

plt.xlabel('EAST')
plt.ylabel('NORTH')
plt.show()
```

```python
# Your start and goal location defined above
# will not necessarily be on the skeleton so you
# must firstidentify the nearest cell on the 
# skeleton to start and goal

def find_start_goal(skel, start, goal):
    skel_coordinates = np.array(np.transpose(np.nonzero(skel)))
    start_nearist_index = np.argmin(np.linalg.norm(skel_coordinates - np.array(start), axis=1, ord=2))
    goal_nearist_index = np.argmin(np.linalg.norm(skel_coordinates - np.array(goal), axis=1, ord=2))
    
    near_start = tuple(skel_coordinates[start_nearist_index])
    near_goal = tuple(skel_coordinates[goal_nearist_index])
    return near_start, near_goal

skel_start, skel_goal = find_start_goal(skeleton, start_ne, goal_ne)

print(start_ne, goal_ne)
print(skel_start, skel_goal)
```

```python
def heuristic_func(position, goal_position):
    return np.linalg.norm(np.array(position) - np.array(goal_position), ord=2)
```
